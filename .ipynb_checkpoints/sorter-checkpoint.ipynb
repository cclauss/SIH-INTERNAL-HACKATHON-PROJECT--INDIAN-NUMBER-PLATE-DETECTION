```python
import os, shutil
```

```python
images=[]
base_location_for_images=''
for base_path, _, image in os.walk('all data/bikes_ocr_new'):
    images=image
    base_location_for_images=base_path
```

```python
training_xml=[]
base_location_for_training_xml=''
for base, _, training in os.walk('all data/labelled/train'):
    base_location_for_training_xml=base
    training_xml=training
```

```python
base_location_for_images
```

```python
len(training_xml)
```

```python
len(images)
```

```python
for i in range(len(images)):
    images[i]=images[i].split('.')[0].strip()
for i in range(len(training_xml)):
    training_xml[i]=training_xml[i].split('.')[0].strip()
```

```python
training_xml
```

```python
for image in images:
    if image in training_xml:
        full_file_path=base_location_for_images+'/'+image+'.jpg'
        shutil.move(full_file_path, 'all data/train_images')
```
